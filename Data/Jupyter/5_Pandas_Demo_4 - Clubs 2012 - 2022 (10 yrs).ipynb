```python
LEAGUES = [
        {"premier_league" : 
                         {"name" : "Premier-League", "code" : 9}
        },
        {"ligue_1" : 
                         {"name" : "Ligue-1", "code" : 13}
        },
        {"bundes_liga" : 
                         {"name" : "Bundesliga", "code" : 20}        
        },
        {"serie_a" :
                         {"name" : "Serie-A", "code" : 11}        
        },
        {"la_liga" :
                         {"name" : "La-Liga", "code" : 12}        
        }
        ]

league_list=[value['name'] for league in LEAGUES for value in league.values()]

league_list
```

```python
import pandas as pd

club_dict={}

for league in league_list:
    df_clubs = pd.read_csv(f'C:/Users/ssm47/Desktop/Git (Projets Perso)/Football/Data/Csv/Bs4_results/Clubs/{league}/{league}_Clubs.csv')
         
    club_name=df_clubs[['Club']]
    club_param=df_clubs[['Club_name','Club_id']]

    club_param_dict=club_param.to_dict('records')
    club_name_dict=club_name.to_dict()
    
    for num in range(0, len(club_name_dict['Club'])):
        club_dict[club_name_dict['Club'][num]] = club_param_dict[num]
        club_dict.update(club_dict)
club_dict
```

```python
import datetime

def getSeason(year):
    season = f"{year}-{year+1}"
    return season
    
today = datetime.datetime.now()

# Year : today.year(2022) By Default
year = 2021
season = getSeason(year)

# premier_league, ligue_1, bundes_liga, serie_a, la_liga
enter_club = "Real Madrid"

def find_club_name(enter_club):
    for club in club_dict:     
        if club == enter_club:
            return club_dict[enter_club]['Club_name']

def find_club_id(enter_club):
    for club in club_dict:     
        if club == enter_club:
            return club_dict[enter_club]['Club_id']

def all_clubs_list():
    clubs=[]
    for club in club_dict:
        clubs.append(club)
    return clubs
        
        
club_name = find_club_name(enter_club)
club_id = find_club_id(enter_club)

print(club_name, club_id)

all_clubs_list()
```

```python
import requests
from bs4 import BeautifulSoup

URL_BASE_PATH = "https://fbref.com/en"

url = f"{URL_BASE_PATH}/squads/{club_id}/{season}/{club_name}-Stats"
res = requests.get(url)
res.raise_for_status()

soup = BeautifulSoup(res.text, 'lxml')
```

```python
all_stats=soup.find(attrs={'id':"all_stats_standard"})

table = all_stats.tbody


players = table.find_all(attrs={'data-stat':'player'})

nationalities = table.find_all(attrs={'data-stat':'nationality'})

for i in players:
    print(i.get_text(), i.a['href'])
```
