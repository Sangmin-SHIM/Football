```python
import requests
from bs4 import BeautifulSoup
from PIL import Image

base_url = "https://www.countryflags.com/"
```

```python
res = requests.get(base_url)
res.raise_for_status()

soup = BeautifulSoup(res.text, 'lxml')
    
# Find a table by id     
tiles = soup.find(attrs={"class":"tiles"})

thumbs = tiles.findAll(attrs={"class":"thumb"})

for thumb in thumbs:
    image_src = thumb.img['src']
    image = Image.open(requests.get(image_src, stream = True).raw)
    if image.mode != 'RGB':
        image = image.convert('RGB')
    image.save(f'C:/Users/ssm47/Desktop/Git (Projets Perso)/Football/Utility/Countries Flag/{thumb.get_text().lstrip("Flag of")}.jpg')
```
