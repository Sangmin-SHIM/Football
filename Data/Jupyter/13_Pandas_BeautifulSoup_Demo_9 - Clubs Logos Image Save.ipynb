```python
import pandas as pd
import re
import os
import csv
import random
import time

# Original File
FILE_BASE_PATH="C:/Users/ssm47/Desktop/Git (Projets Perso)/Football/Data/Csv/Bs4_results/Clubs/All-Clubs"
clubs_with_id_file = "All-Clubs_with_Club_Id"

df_club=pd.read_csv(f'{FILE_BASE_PATH}/{clubs_with_id_file}.csv', sep=";")
```

```python
club_link_list=[df_club.loc[num].to_list()[1] for num in range (0,len(df_club))]
club_name_list=[df_club.loc[num].to_list()[3] for num in range (0,len(df_club))]

club_link_list
```

```python
import csv
import requests
from bs4 import BeautifulSoup
from PIL import Image

URL_BASE_PATH = "https://fbref.com/en"

for num in range(0, len(df_club)):

    # BeautifulSoup     
    url = f"{URL_BASE_PATH}/{club_link_list[num]}"
    res = requests.get(url)
    res.raise_for_status()
    soup = BeautifulSoup(res.text, 'lxml')

    
    image_src = soup.find(attrs={"class":"teamlogo"})['src']
    image = Image.open(requests.get(image_src, stream = True).raw)
    if image.mode != 'RGB':
        image = image.convert('RGB')
    image.save(f'C:/Users/ssm47/Desktop/Git (Projets Perso)/Football/Utility/Clubs Logo/{club_name_list[num]}.png')

    print(f"{club_name_list[num]} DONE ! --- {num} / {len(df_club)}")
    
    time.sleep(random.randint(2,8))
```
