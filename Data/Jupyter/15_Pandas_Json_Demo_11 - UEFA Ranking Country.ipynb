```python
import requests
import pandas as pd
import csv


def csv_write(contents):
    writer.writerow(contents)

def data_json(url,year):
    url = UEFA_BASE_URL
    params={"coefficientType" : "MEN_ASSOCIATION",
        "coefficientRange" : "OVERALL",
        "seasonYear" : year,
        "page" : "1",
        "pagesize" : "500",
        "language" :"en"
       }
    headers = {'Accept': 'application/json'}
    res = requests.get(UEFA_BASE_URL, headers=headers, params=params)
    
    return res.json()['data']['members']

def get_detail_data(countries):
    for index in range(0,len(countries)):
        # VARIABLE
        country=countries[index]
        
        #-------------------DATA---------------------
        countryName= country['member']['countryName']
        position=country['overallRanking']['position']
        year=country['overallRanking']['targetSeasonYear']
        points=country['overallRanking']['totalPoints']
        nbTeams=country['seasonRankings'][0]['numberOfTeams']
        nbMatches=country['seasonRankings'][0]['numberOfMatches']
        
        
        # VARIABLES
        seasons_list=country['seasonRankings']
        based_years = {}
        for num in range(1,len(seasons_list)):
            based_years[year-num] = country['seasonRankings'][num]["totalPoints"]
        
        contents = [year, countryName, position, points, nbMatches, nbTeams, *based_years.values()]
        csv_write(contents)
        print(index+1, countryName, position, year, points, nbMatches, nbTeams, *based_years.values())
        #-------------------/DATA--------------------
```

```python
years= [year for year in range(2002,2023)]
UEFA_BASE_URL = "https://comp.uefa.com/v2/coefficients"

#Generate Csv File
new_file = open(f'test2.csv', 'w', newline='', encoding='utf=8')
writer = csv.writer(new_file)
headers = ["year","countryName", "position",  "points", "nbMatches", "nbTeams", "year_1(based)", "year_2(based)", "year_3(based)", "year_4(based)"]
writer.writerow(headers)


for year in years:
    countries = data_json(UEFA_BASE_URL, year)
    get_detail_data(countries)
new_file.close()
```
